# 基礎編5
このノートブックでは、現在実行中のトランザクションの情報を取得するスマートコントラクトAPIを使用する簡単な例を示します。

- 使用するスマートコントラクトAPI:   
getTxno, getTxid, getCallerId, getContractId, getContractDomainId, getCallerDomainId


## APIの説明 

getTxno(): 実行しているトランザクションの番号を返す。   
getTxid(): 実行しているトランザクションのTXIDを返す。  
getCallerId(): 実行しているトランザクションを呼び出したユーザまたはコントラクトのIDを返す。  
getContractId(): 実行しているトランザクションのコントラクトのIDを返す。  
getCallerDomainId(): 実行しているトランザクションを呼び出したユーザまたはコントラクトが所属するドメインのIDを返す。  
getContractDomainId(): 実行しているトランザクションのコントラクトが所属するドメインのIDを返す。

## 準備
実行の準備を行います。

```javascript
var { adminWallet, rpc, deploySmartContract } = await import('../lib/notebook-util.mjs');
```

## スマートコントラクトのデプロイ

現在実行中のトランザクションの情報を返す簡単なスマートコントラクトを作成します。

```javascript
var cid = await deploySmartContract(function basic5(){
    return {
        txno: getTxno(),
        txid: getTxid(),
        caller: getCallerId(),
        callee: getContractId(),
        callerDomain: getCallerDomainId(),
        calleeDomain: getContractDomainId()
    };  
});
```

## 実行

```javascript
var resp = await rpc.call(adminWallet, cid);
console.log(resp);
```

## 確認

c1queryで取得したトランザクション記録と、上記APIで得られた値が整合するかを確認します。

```javascript
var { txno, txid, addr, caller, callee } = resp.value;
var { value } = await rpc.call(adminWallet, 'c1query', { type: 'a_transaction', id: resp.txno });
console.log(value);
if (txno == value.txno) console.log('OK'); else console.log('NG');
if (txid == value.txid) console.log('OK'); else console.log('NG');
if (caller == value.caller[0]) console.log('OK'); else console.log('NG');
if (callee == value.callee[0]) console.log('OK'); else console.log('NG');
```
