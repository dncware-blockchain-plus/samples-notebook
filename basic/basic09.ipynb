# 基礎編9
このノートブックでは、UnifiedNameに関する例を示します。

## UnifiedNameの特徴

- UnifiedNameはobjectname@domainnameの形式の文字列です。
- ブロックチェーン内で一意です。
- オブジェクトの作成時に名前を指定することができ、あとから変更することもできます。（例外あり）

一方、比較として、オブジェクトのIDには以下のような特徴があります。
- IDは「英数字1文字＋数字のならび」という形式の文字列です。
- ブロックチェーン内で一意です。
- オブジェクトの作成時に自動的に採番され、変更することはできません。

## 準備

```javascript
var { adminWallet, rpc, deploySmartContract } = await import('../lib/notebook-util.mjs');
```

## 呼び出した自分自身のIDとUnifiedNameを問い合わせる

c1query-serachのkeyに'me'を指定して問い合わせます。  
IDとUnifiedNameが返ります。

```javascript
var resp = await rpc.call(adminWallet, 'c1query', { type: 'search', key: 'me' });
console.log(resp);
var uid = resp.value[0].id;
var uname = resp.value[0].name;
```

## UnifiedNameからIDを問い合わせる

c1query-serachのkeyにUnifiedNameを指定して問い合わせます。

```javascript
var resp = await rpc.call(adminWallet, 'c1query', { type: 'search', key: uname });
console.log(resp);
```

## IDからUnifiedNameを問い合わせる

c1query-serachのkeyにIDを指定して問い合わせます。

```javascript
var resp = await rpc.call(adminWallet, 'c1query', { type: 'search', key: uid });
console.log(resp);
```

## コントラクトの中で問い合わせる

getIdByUnifiedNameを使う簡単なコントラクトを作成します。

```javascript
var cid = await deploySmartContract({ uname: 'string' }, function basic9(uname){
    var id = getIdByUnifiedName(uname);
    return id;
});
```

作成したコントラクトを呼び出します。

```javascript
var resp = await rpc.call(adminWallet, cid, { uname });
console.log(resp);
```

## IDのかわりにUnifiedNameを使う

ほとんどのAPIは、IDのかわりにUnifiedNameを受け付けることができます。  
ここでは、rpc.call呼び出しの第２引数のコントラクトIDのところでUnifiedNameを指定します。

```javascript
var resp = await rpc.call(adminWallet, 'c1query', { type: 'search', key: cid });
console.log(resp);
var contract_uname = resp.value[0].name;
var resp = await rpc.call(adminWallet, contract_uname, { uname });
console.log(resp);
```
