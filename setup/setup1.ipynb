# 初期セットアップ（ステップ１）
このノートブックでは、サンプルコードの実行環境を初めて使用する際の設定のステップ１を行います。  
ステップ１では、ウォレットの作成を行います。  
ウォレットファイルは実行環境のetcフォルダに格納されます。（ウォレットファイルが存在する場合は上書きされません）

## 必要なライブラリの読み込み

```javascript
var { api } = await import('../lib/load-blockchain-api.mjs');
```

## ウォレットの作成
ブロックチェーンのウォレットを生成します。

パスワードをランダムに生成します。

```javascript
var password = api.makeRandomText(8);
```

ウォレットのJSONを生成します。

```javascript
var walletJSON = await api.createWalletFile('admin-wallet', password, 'es');
```

ウォレットを開いて、そのアドレスを表示します。

```javascript
var wallet = await api.unlockWalletFile(await api.parseWalletFile(walletJSON), password);
var address = wallet.address;
console.log('wallet address:', address);
```

## ファイルの保存
ウォレットファイルをetcフォルダに書き込みます。あわせてパスワードを設定ファイルに書き込みます。  
すでにウォレットファイルが存在する場合には上書きしません。(あやまって実行した時に上書きされない)  

```javascript
var fs = await import('node:fs');
var path = await import('node:path');
var { default: package_root } = await import('../lib/get-package-root.mjs');
var wallet_filename = path.join(package_root, 'etc', 'admin-wallet.json');
var config_filename = path.join(package_root, 'etc', 'config.json');
if(fs.existsSync(wallet_filename)) throw 'ウォレットファイルがすでに存在しています: '+wallet_filename;
if(fs.existsSync(config_filename)) {
    var config = JSON.parse(fs.readFileSync(config_filename, 'utf8'));
} else {
    do {
        var uniqueID = api.makeRandomText(6);
    } while(/\d/.test(uniqueID[0]));
    var config = { uniqueID };
}
config.address = address;
config.password = password;
fs.writeFileSync(config_filename, JSON.stringify(config, null, 1), 'utf8');
fs.writeFileSync(wallet_filename, walletJSON, 'utf8');
console.log('DONE');
```
