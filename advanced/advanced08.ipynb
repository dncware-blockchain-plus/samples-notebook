# 応用編8

V3.3新機能:スマートコントラクトの引数型の拡張  

このノートブックでは、スマートコントラクト・バージョン3.3から利用できる、スマートコントラクトの引数型の拡張について説明します。  


## スマートコントラクトの引数型の概要

スマートコントラクトの引数型は、スマートコントラクトに渡された引数の型検査を行う際に使用されます。  
引数型には、基本引数型、配列引数型、マルチ引数型の３種類があります。  
基本引数型は文字列で表現され、'string' 'number' 'boolean' 'json' 'any' 'integer' 'address' 'id' 'base64'のいずれかです。  
配列引数型は文字列で表現され、いずれかの基本引数型の後ろに'-array'をつなげたものです。その基本引数型の配列となっているときに型検査を通過する仕様です。  
マルチ引数型は基本引数型または配列引数型の配列で表現され、リストされたいずれかの引数型に適合するとき型検査を通過する仕様です。

なお、V3.3より前の引数型の仕様は下記の通りでした。
- 基本引数型は、'string' 'number' 'boolean' 'json' 'any' のいずれか。
- 配列引数型、マルチ引数型はサポートされない。

○基本引数型の意味
- string: 文字列
- number: 数値
- boolean: trueまたはfalse
- json: JSON文字列
- any: 任意の値
- integer: 整数
- address: api.isValidWalletAddressFormat()がtrueになる文字列
- id: 有効なObjectIdまたは有効なUnifiedName
- base64: Base64エンコードされた文字列

## 準備

```javascript
var { domain } = await import('../lib/load-config.mjs');
var { adminWallet, rpc, deploySmartContract } = await import('../lib/notebook-util.mjs');
```

説明用のスマートコントラクトをデプロイします。  
このコントラクトは、引数aの値をそのまま返します。  

```javascript
var cid = await deploySmartContract({ a: 'any' }, function adv8(a){
    return a;
});
```

## integer引数型

引数aをinteger引数型として設定します。

```javascript
var resp = await rpc.call(adminWallet, 'c1update', { id: cid, prop: 'argtypes', value: { a: 'integer' } });
console.log(resp);
```

aに123を指定してコントラクトを呼び出します。  
123がそのまま返ります。

```javascript
var resp = await rpc.call(adminWallet, cid, { a: 123 });
console.log(resp);
```

aに0.5を指定してコントラクトを呼び出します。  
引数が整数でないためエラーとなります。

```javascript
var resp = await rpc.call(adminWallet, cid, { a: 0.5 });
console.log(resp);
```

## address引数型

引数aをaddress引数型として設定します。

```javascript
var resp = await rpc.call(adminWallet, 'c1update', { id: cid, prop: 'argtypes', value: { a: 'address' } });
console.log(resp);
```

aにウォレットアドレス文字列を指定してコントラクトを呼び出します。  
（ウォレットアドレスがブロックチェーンに登録されている必要はありません）  
実行すると、アドレス文字列がそのまま返ります。

```javascript
var resp = await rpc.call(adminWallet, cid, { a: 'egwMgRNExy8sJnanwTmKL3FoF3w9uk' });
console.log(resp);
```

aにウォレットアドレス形式ではない文字列を指定してコントラクトを呼び出します。  
実行すると、型検査でエラーとなります。

```javascript
var resp = await rpc.call(adminWallet, cid, { a: '123456789' });
console.log(resp);
```

## id引数型

引数aをid引数型として設定します。

```javascript
var resp = await rpc.call(adminWallet, 'c1update', { id: cid, prop: 'argtypes', value: { a: 'id' } });
console.log(resp);
```

aにIDを指定してコントラクトを呼び出します。  
実行すると、IDがそのまま返ります。  

```javascript
var resp = await rpc.call(adminWallet, cid, { a: cid });
console.log('cid:', cid);
console.log(resp);
```

aにUnifiedName文字列を指定してコントラクトを呼び出します。  
実行すると、UnifiedNameが指すオブジェクトのIDが返ります。  
（引数型のチェック時に名前解決が行われ、結果のIDが引数としてコードに渡されます）

```javascript
var uname = 'adv8@'+domain; // 説明用コントラクトのUnifiedName
console.log(uname);
var resp = await rpc.call(adminWallet, cid, { a: uname });
console.log(resp);
```

aにブロックチェーンに登録されていないIDを指定してコントラクトを呼び出します。  
実行すると、型検査でエラーとなります。

```javascript
var resp = await rpc.call(adminWallet, cid, { a: 'u1234567' });
console.log(resp);
```

## base64引数型

引数aをbase64引数型として設定します。

```javascript
var resp = await rpc.call(adminWallet, 'c1update', { id: cid, prop: 'argtypes', value: { a: 'base64' } });
console.log(resp);
```

aにBase64エンコードされた文字列を指定してコントラクトを呼び出します。  
実行すると、Base64文字列がそのまま返ります。

```javascript
var resp = await rpc.call(adminWallet, cid, { a: 'QUJDREVGRw==' });
console.log(resp);
```

aにBase64文字列でない文字列を指定してコントラクトを呼び出します。  
実行すると、型検査でエラーとなります。

```javascript
var resp = await rpc.call(adminWallet, cid, { a: '...' });
console.log(resp);
```

## 配列引数型

引数aをinteger-array引数型として設定します。

```javascript
var resp = await rpc.call(adminWallet, 'c1update', { id: cid, prop: 'argtypes', value: { a: 'integer-array' } });
console.log(resp);
```

aに整数の配列を指定してコントラクトを呼び出します。  
実行すると、その配列がそのまま返ります。

```javascript
var resp = await rpc.call(adminWallet, cid, { a: [ 1, 0, -123 ] });
console.log(resp);
```

aに整数を指定してコントラクトを呼び出します。  
実行すると、配列ではないためエラーとなります。

```javascript
var resp = await rpc.call(adminWallet, cid, { a: 123 });
console.log(resp);
```

aに整数でない値を含む配列を指定してコントラクトを呼び出します。  
実行すると、型検査でエラーとなります。

```javascript
var resp = await rpc.call(adminWallet, cid, { a: [ 1, 2, null ] });
console.log(resp);
```

## マルチ引数型

引数aをintegerとbase64のマルチ引数型として設定します。

```javascript
var resp = await rpc.call(adminWallet, 'c1update', { id: cid, prop: 'argtypes', value: { a: [ 'integer', 'base64' ] } });
console.log(resp);
```

aに整数を指定してコントラクトを呼び出します。  
実行すると、そのままの値が返ります。

```javascript
var resp = await rpc.call(adminWallet, cid, { a: 123 });
console.log(resp);
```

aにBase64文字列を指定してコントラクトを呼び出します。  
実行すると、そのままの値が返ります。

```javascript
var resp = await rpc.call(adminWallet, cid, { a: 'QUJDREVGRw==' });
console.log(resp);
```

aに整数でもBase64文字列でもない値を指定してコントラクトを呼び出します。  
実行すると、型検査でエラーとなります。

```javascript
var resp = await rpc.call(adminWallet, cid, { a: '-123' });
console.log(resp);
```
