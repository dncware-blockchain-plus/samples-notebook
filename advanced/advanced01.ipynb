# 応用編1
このノートブックでは、トランザクション要求を作成し署名する方法を示します。

## 概要
rpc.callを使ったトランザクションの呼び出しでは、トランザクション要求の作成/署名は内部処理として自動的に行われますが、  
本来は、トランザクション要求の作成/署名とブロックチェーンへの送信を分けて行うこともできます。  
たとえば、ネットワークに接続できない場所でトランザクション要求に署名し、のちにネットワークにつながった時に送信する場合などで利用します。

## 準備

実行の準備を行います。

```javascript
var { api } = await import('../lib/load-blockchain-api.mjs');
var { chainID, domain } = await import('../lib/load-config.mjs');
var { adminWallet, rpc, deploySmartContract } = await import('../lib/notebook-util.mjs');
```

テスト用の簡単なコントラクトをデプロイします。

```javascript
var cnt = await deploySmartContract(function adv1() {
    return new Date();
});
```

## トランザクション要求の作成と署名

トランザクション要求を作成します。  

```javascript
var request = await api.createRequest(adminWallet.address, cnt, {});
console.log(request);
```

トランザクション要求に署名します。

```javascript
await api.signRequest(request, adminWallet, chainID);
console.log(request);
```

## トランザクション要求を送信

ここでは、rpc._call_requestを使って送信する例を示します。

```javascript
var resp = await rpc._call_request(request);
console.log(resp);
```

## おなじトランザクション要求を再送信

再送した場合、まったく同じ結果が返ってきます。  

```javascript
var resp = await rpc._call_request(request);
console.log(resp);
```
