# 応用編9

V3.5新機能:ピアの構成情報の取得

このノートブックでは、ピアの構成情報を取得する方法を説明します。


## 準備

```javascript
var { api } = await import('../lib/load-blockchain-api.mjs');
var { rpc } = await import('../lib/notebook-util.mjs');
```

## ブロックチェーンからピア構成の最新番号を取得する

```javascript
var { V } = await rpc.fetchCnfstr();
console.log(V);
```

## ピア構成情報をブロックチェーンから取得する

最新のピア構成情報をブロックチェーンから取得します。  
取得したピア構成情報(cnfstr)はJSON文字列で、各ピアのIDや公開鍵、アクセスポイントのURLなどが含まれます。

```javascript
var { cnfstr } = await rpc.fetchCnfstr(V);
var cnf = JSON.parse(cnfstr);
console.log(cnf);
```

## ピア構成情報オブジェクトへの変換

cnfstrが所定のフォーマットであるか検査し、オブジェクトに読み込みます。

```javascript
var peerscnf = await api.loadPeersCnf(cnfstr);
```

## ピア構成情報オブジェクトからの情報取得

pidの一覧を配列として取得できます。

```javascript
var pids = peerscnf.pids;
console.log(pids);
```

オーソリティピアに限ったpidの一覧を配列として取得できます。

```javascript
var authorities = peerscnf.authorities;
console.log(authorities);
```

特定のピアの情報を取得できます。  
ここで、peerscnf.peersは、ピアのIDをキーとするMapオブジェクトです。

```javascript
var peer = peerscnf.peers.get(authorities[0]);
console.log(peer);
```

たとえば、各ピアのアクセスポイントのURLを列挙したい場合には、以下のようにします。

```javascript
for(var [pid, peer] of peerscnf.peers){
    var url = peer.url;
    if(url) console.log(pid, url);
}
```
