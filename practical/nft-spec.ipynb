## NFTコントラクトの動作
NFTコントラクトの動作は以下の通りとします。

NFTコントラクトの呼び出しでは、funcに機能名を指定し、argsにその機能に関する引数を指定します。  
NFTコントラクトは、funcで指定された機能をargsで指定された引数に応じて実行し、戻り値を返します。

## onNFTReceivedの定義
NFTコントラクトは、safeTransferFromの実行に際して、トークンの転送先コントラクトがonNFTReceivedに応答するかを確認します。
- onNFTReceivedは、トークンの転送を行った後の状態で呼び出されます。
- onNFTReceivedが正常に実行され'd56Kq6n1'が返った場合には、トランザクションは成功し、トークンは転送されます。
- そうでない場合には、トランザクションはアボートし、トークンは転送されません。（元の状態にロールバックする）

なお'd56Kq6n1'は、sha256(utf8encode('onNFTReceived(operator,from,tokenId,data)')).substring(0,8)により求められます。  

NFTコントラクトからの呼び出しの具体的なコードは以下の通りとします。
```
openContract(to) // トークンの転送先のコントラクト
.call({
    func: 'onNFTReceived',
    args: {
        operator: caller, // safeTransactionFromを呼び出したcaller
        from: from, // 直近のオーナ(=safeTransferFromに渡されたargs.from)
        tokenID: tokenID, // 転送しようとしているトークン
        data: data, // safeTransferFromに渡されたargs.data
    }
})
```

## 特記事項
* トークンのIDは英数字（上限100文字まで）で構成される文字列とします。
* トークンのIDはNFTコレクション内で一意であるとします。
* トークンのIDは人が識別しやすい文字列であることが望ましいです。
* トークンのオーナは、ユーザまたはコントラクトのいずれかです。anonymous、ウォレット、システムコントラクト、Key-valueストアはオーナにはなりません。
* NFTコントラクトは、管理者による変更不可・削除不可とすることが望ましいです。
* transferFrom/safeTransferFromの処理の一環として`discloseTo(from,to)`を実行することが望ましいです。（取引当事者にトランザクションログを開示するため）
* ownerOfとgetApprovedの返り値の文字列はID形式(英数字1文字のあとに数字が続く形式）です。
* NFTコントラクトへのanonymousによるアクセスは一律にエラーとします。

## 独自の機能拡張
以下に該当する場合、この文書で定められていない独自の機能をNFTコントラクトに実装することができます。
* NFTコントラクトの引数型に引数を追加しても良いです。
* funcで指定される、NFTコントラクトの各機能を増やしても良いです。例えばトークンを生成するfuncとして'mint'を追加できます。
* 戻り値が「なし」の場合、独自にどのような値を返しても良いです。
