# DNCWARE Blockchain+におけるコインインタフェース定義（ドラフト版）

## 概要 
イーサリアムのERC20トークンに相当するスマートコントラクトのインタフェースを定義します。

なお、１つのスマートコントラクトで１種類のコインを実現します。  

## コインコントラクトの動作
コインコントラクトの動作は以下の通りとします。

コインコントラクトの呼び出しでは、funcに機能名を指定し、argsにその機能に関する引数を指定します。  
コインコントラクトは、funcで指定された機能をargsで指定された引数に応じて実行し、戻り値を返します。

### approve

転送は複数回に分けて行うことも可能です。  
approveが再度呼ばれた場合には、転送の上限数は上書きされます。  
oldが指定される場合、上書きされる値がoldと一致しないとエラーとなります。

## 特記事項
* コインのオーナは、ユーザまたはコントラクトのいずれかです。anonymous、ウォレット、システムコントラクト、Key-valueストアはオーナにはなりません。
* コインコントラクトは、管理者による変更不可・削除不可とすることが望ましいです。
* transfer/transferFromの処理の一環として`discloseTo(from,to)`を実行することが望ましいです。（取引当事者にトランザクションログを開示するため）
* コインコントラクトへのanonymousによるアクセスは一律にエラーとします。

## 独自の機能拡張
以下に該当する場合、この文書で定められていない独自の機能をコインコントラクトに実装することができます。
* コインコントラクトの引数型に引数を追加しても良いです。
* funcで指定される、コインコントラクトの各機能を増やしても良いです。
* 戻り値が「なし」の場合、独自にどのような値を返しても良いです。
